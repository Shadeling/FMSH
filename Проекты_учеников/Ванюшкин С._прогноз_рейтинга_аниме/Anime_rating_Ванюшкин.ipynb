```python
data= data.dropna(axis=0, how='any')
data.reset_index(drop=True, inplace=True)
print(data.shape)
data
```

```python
a=[]
numberep=[]
for num in data['episodes']:
    numberep.append(num)
numberlist=set(numberep)
print(numberlist)
len(numberlist) 
```

```python
data = data[lambda x: x['episodes'] != 'Unknown']
data.reset_index(drop=True, inplace=True)
data
```

```python
data['episodes'] = pd.to_numeric(data['episodes'])
data.dtypes
```

```python
dataDummies = pd.get_dummies(data['type'], prefix = 'type')
dataDummies.head()
```

```python
data = pd.concat([data, dataDummies], axis=1)
data
```

```python
del data['type']
data.head()
```

```python
a=[]
genreanime=[]
for genre in data['genre']:
    spliter = genre.split(",")
    for val in spliter:
        val = val.strip()
        genreanime.append(val)
genrelist=set(genreanime)
print(genrelist)
len(genrelist)  
```

```python
for i in genrelist:
    data[i]=0
```

```python
data.head()
```

```python
i=0
for genreanime in data['genre']:
    for genre in genrelist:
            if genre in genreanime:
                data.loc[[i], genre]=1
            else:
                data.loc[[i], genre]=0
    i=i+1
```

```python
data.head(15)
```

```python
del data['genre']
del data['name']
del data['anime_id']
data.head()
```

```python
sc = MinMaxScaler()
for col in data.columns:
    if col != 'rating':
        data[col] = sc.fit_transform(data[[col]])
```

```python
msk = np.random.rand(len(data)) < 0.8

train=data[msk]
test=data[~msk]

y_train=train['rating']
y_test=test['rating']

del test['rating']
del train['rating']
```

```python
def create_model(init):
    model = keras.Sequential()
    model.add(keras.layers.Dense(100, input_shape=(train.shape[1],), kernel_initializer=init, activation='tanh'))
    model.add(keras.layers.Dense(100, kernel_initializer=init, activation='tanh'))
    model.add(keras.layers.Dense(100, kernel_initializer=init, activation='tanh'))
    model.add(keras.layers.Dense(100, kernel_initializer=init, activation='tanh'))
    model.add(keras.layers.Dense(1, kernel_initializer=init, activation='linear'))
    return model
```

```python
uniform_model = create_model("uniform")
uniform_model.compile(
loss='mean_absolute_error', optimizer=keras.optimizers.SGD(lr=0.1), metrics=['mean_absolute_percentage_error'])
uniform_model.fit(train, y_train,
batch_size=64, nb_epoch=1000, verbose=2, validation_data=(test, y_test))
```
